### üìö Importaci√≥n de librer√≠as
Se importa OpenCV (`cv2`) para procesamiento de im√°genes y `cv2_imshow` para mostrar im√°genes en Google Colab.

````python
import cv2
from google.colab.patches import cv2_imshow
````

### üñº Lectura y visualizaci√≥n de la imagen
Se carga la imagen en formato BGR, se muestra su tama√±o, tipo de dato y memoria utilizada. Luego, se visualiza.

````python
imagen_bgr = cv2.imread('frutos_rojos.png')
print("Dimensiones de la imagen:", imagen_bgr.shape)
print("Tipo de datos:", imagen_bgr.dtype)
print("Tama√±o en memoria (bytes):", imagen_bgr.nbytes)
cv2_imshow(imagen_bgr)
````

### üîµ Canal Azul (B)
Se extrae el canal azul de la imagen (√≠ndice 0), se calculan estad√≠sticas b√°sicas y se muestra el canal.

````python
B = imagen_bgr[:,:,0]
print("Estad√≠sticas del canal Azul:")
print(f"Valor m√≠nimo: {B.min()}, Valor m√°ximo: {B.max()}, Promedio: {B.mean():.2f}")
cv2_imshow(B)
````

### üü¢ Canal Verde (G)
Se extrae el canal verde (√≠ndice 1), se muestran estad√≠sticas y su visualizaci√≥n.

````python
G = imagen_bgr[:,:,1]
print("\nEstad√≠sticas del canal Verde:")
print(f"Valor m√≠nimo: {G.min()}, Valor m√°ximo: {G.max()}, Promedio: {G.mean():.2f}")
cv2_imshow(G)
````

### üî¥ Canal Rojo (R)
Se analiza el canal rojo (√≠ndice 2) y se presentan sus caracter√≠sticas.

````python
R = imagen_bgr[:,:,2]
print("\nEstad√≠sticas del canal Rojo:")
print(f"Valor m√≠nimo: {R.min()}, Valor m√°ximo: {R.max()}, Promedio: {R.mean():.2f}")
cv2_imshow(R)
````

### üìä Importaci√≥n de Matplotlib
Se importa la biblioteca `matplotlib.pyplot` para una visualizaci√≥n alternativa.

### üñº Visualizaci√≥n con Matplotlib
Se carga la imagen en formato RGB y se muestra con Matplotlib, incluyendo t√≠tulo, ejes y barra de color.

````python
imagen_rgb = plt.imread('frutos_rojos.png')
plt.figure(figsize=(10,8))
plt.imshow(imagen_rgb)
plt.title('Imagen RGB en Matplotlib')
plt.axis('on')
plt.colorbar()
plt.show()
````

### üé® ¬øQu√© significa la **barra de color** en `plt.colorbar()`?

Cuando en una imagen us√°s `plt.imshow()` y luego `plt.colorbar()`, **se muestra una barra lateral con una escala de colores**. Esta barra indica qu√© **valores num√©ricos** (por ejemplo, intensidades de color o niveles de gris) est√°n representados por qu√© **colores** en la imagen.

---

### üëá En este contexto:

```python
imagen_rgb = plt.imread('frutos_rojos.png')
plt.imshow(imagen_rgb)
plt.colorbar()
```

üìå Lo que est√° pasando:

- `plt.imread()` carga la imagen como una **matriz NumPy**, donde cada p√≠xel tiene valores entre 0 y 1 (si es RGB normalizado).
- `plt.imshow()` interpreta esos valores y los representa visualmente.
- `plt.colorbar()` agrega una **escala lateral** que muestra c√≥mo se relacionan los **valores num√©ricos** de los p√≠xeles con los colores.

---

### üß† Pero ojo con esto:

- Si la imagen es en color RGB como en tu notebook, **la barra de color puede no ser muy √∫til**, porque cada p√≠xel tiene 3 valores (R, G, B), y la escala de colores de la barra solo representa **un valor por p√≠xel**.
- Por eso, la barra de color **es m√°s √∫til cuando mostr√°s una sola capa**, como una imagen en escala de grises o una m√°scara binaria.

### üì∑ Ejemplo pr√°ctico:

Si mostraras solo un canal, por ejemplo el azul:

````python
plt.imshow(B, cmap='gray')
plt.colorbar()
````



Ah√≠ s√≠ la **barra de color tiene mucho sentido**:
- Te muestra que el valor `0` se ve negro.
- El valor `255` se ve blanco.
- Y los valores intermedios son tonos de gris.

---

### ‚úÖ En resumen:
La barra de color es una **leyenda visual** que **traduce valores num√©ricos a colores**. Sirve sobre todo cuando mostr√°s:
- Im√°genes en escala de grises
- Mapas de calor
- M√°scaras de segmentaci√≥n
